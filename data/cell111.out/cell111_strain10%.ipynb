```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("energias.dat", 'r') as f:
    dados = f.readlines()
dados = [valor.split(" ") for valor in dados]
dados = np.array(dados)
filename, energias = dados[:,0], dados[:,1]
energias = [float(valor.replace(r'\n','')) for valor in energias]
```

```python
fig, ax = plt.subplots(figsize=(7,5))
ax.set_title("Energias em Ry")
ax.scatter(x=filename, y=energias)
ax.set_ylabel("Energia")
ax.set_xlabel("filename")
ax.set_xticks([])
# ax.set_xticks()
plt.tight_layout()
```
