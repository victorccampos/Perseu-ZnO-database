```python
import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np
%matplotlib inline
```

```python

plt.rcParams["figure.dpi"]=150
plt.rcParams["figure.facecolor"]="white"
plt.rcParams["figure.figsize"]=(8, 6)
```

```python

# load data
data = np.loadtxt('ZnO-fixed_bands.dat.gnu')
```

```python

k = np.unique(data[:, 0])
bands = np.reshape(data[:, 1], (-1, len(k)))
```

```python
sim_points =[0.0000,0.5774,
0.8871,
1.4644,
1.7741,
2.7924,
3.1021,
4.1204]

ZB_points = ['$\Gamma$','K','L', 'A', '$\Gamma$', 'K', 'H', 'A']
```

```python
# Plot das bandas (como j est no seu cdigo)
for band in range(len(bands)):
    plt.plot(k, bands[band, :], linewidth=1, color='k')
plt.xlim(min(k), max(k))

# Linhas horizontais de referncia (nveis de energia)
plt.axhline(6.7701, linestyle=(0, (5, 5)), linewidth=0.75, color='k', alpha=0.5)
plt.axhline(7.5097, linestyle=(0, (5, 5)), linewidth=0.75, color='k', alpha=0.5)

# Pontos de alta simetria
plt.axvline(0.0000, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(0.5774 , linewidth=0.75, color='k', alpha=0.5)
plt.axvline(0.8871, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(1.4644, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(1.7741, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(2.7924 , linewidth=0.75, color='k', alpha=0.5)
plt.axvline(3.1021, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(4.1204, linewidth=0.75, color='k', alpha=0.5)

# Labels dos pontos de Brillouin
plt.xticks(ticks=sim_points, labels=ZB_points)

# Adiciona a seta do band gap
x_gap = 1.7441  # ou qualquer posio x que esteja livre
plt.annotate(
    '', xy=(x_gap, 7.5097), xytext=(x_gap, 6.7701),
    arrowprops=dict(arrowstyle='<->', color='red', linewidth=2)
)

# Texto do valor do gap
plt.text(
    x_gap + 0.05, (6.7701 + 7.5097)/2,
    f'{7.5097 - 6.7701:.2f} eV',
    va='center', color='red'
)

plt.ylabel("Energy (eV)")
plt.ylim([0, 12])
plt.show()
```
