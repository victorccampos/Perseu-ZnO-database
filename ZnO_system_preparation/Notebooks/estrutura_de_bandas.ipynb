```python
data = np.loadtxt(r'ZnO_bulk_bands_noGGApU.dat.gnu')
```

```python
with plt.style.context('~/computermodern.mplstyle'):
    k = np.unique(
    data[:,0]
)
    bands = np.reshape(data[:, 1], (-1, len(k)))


    fig, ax = plt.subplots(figsize=(8,6))
    for band in range(len(bands)):
        plt.plot(k, bands[band, :], linewidth=1, alpha=0.5, color='k')

    # high simetry-points
    plt.axvline(0.5774, linewidth=0.75, color='k', alpha=0.5)
    plt.axvline(0.8871, linewidth=0.75, color='k', alpha=0.5)
    plt.axvline(1.4644, linewidth=0.75, color='k', alpha=0.5)
    plt.axvline(1.7741, linewidth=0.75, color='k', alpha=0.5)
    plt.axvline(2.7924, linewidth=0.75, color='k', alpha=0.5)
    plt.axvline(3.1021, linewidth=0.75, color='k', alpha=0.5)
    plt.axvline(4.1204, linewidth=0.75, color='k', alpha=0.5)


    # text labels of high simmetry points
    plt.xticks(ticks= [0, 0.5774, 0.8871, 1.4644, 1.7741, 2.7924, 3.1021, 4.1204], \
            labels=['$\Gamma$', 'M', 'L', 'A', '$\Gamma$', 'K', 'H', 'A'])

    plt.xlim(min(k), max(k))
    plt.ylim(6, 8)
    plt.ylabel("Energy (eV)")

    plt.axhline(6.77, linewidth=0.75, color='red', alpha=0.5)
    plt.axhline(7.51, linewidth=0.75, color='red', alpha=0.5)
    plt.axhline(7.1893, linewidth=1, color='red', label='Fermi Energy')
    plt.legend()

    print(f'Band Gap: {7.51 - 6.77}')
    
```

```python
k = np.unique(
    data[:,0]
)
bands = np.reshape(data[:, 1], (-1, len(k)))


fig, ax = plt.subplots(figsize=(8,6))
for band in range(len(bands)):
    plt.plot(k, bands[band, :], linewidth=1, alpha=0.5, color='k')

# high simetry-points
plt.axvline(0.5774, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(0.8871, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(1.4644, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(1.7741, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(2.7924, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(3.1021, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(4.1204, linewidth=0.75, color='k', alpha=0.5)


# text labels of high simmetry points
plt.xticks(ticks= [0, 0.5774, 0.8871, 1.4644, 1.7741, 2.7924, 3.1021, 4.1204], \
           labels=['$\Gamma$', 'M', 'L', 'A', '$\Gamma$', 'K', 'H', 'A'])

plt.xlim(min(k), max(k))
plt.ylim(6, 8)
plt.ylabel("Energy (eV)")

plt.axhline(6.77, linewidth=0.75, color='red', alpha=0.5)
plt.axhline(7.51, linewidth=0.75, color='red', alpha=0.5)
plt.axhline(7.1893, linewidth=1, color='red', label='Fermi Energy')
plt.legend()

print(f'Band Gap: {7.51 - 6.77}')
```
