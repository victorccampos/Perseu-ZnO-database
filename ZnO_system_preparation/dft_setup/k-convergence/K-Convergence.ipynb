```python
# Criar o gráfico

plt.figure(figsize=(7, 5))
plt.scatter(k, total_energy, color='black', s=50, label='Energia total')

# Linha de referência no maior k.
plt.axhline(total_energy[-1], color='gray', linestyle='--', label=f'Referência (k={k[-1]})')

# Rótulos e limites
plt.suptitle('Convergência Grid $k$-pontos', fontsize=14)
plt.xlabel('Número de k-pontos',fontsize=12)
plt.ylabel('Energia total (Ry)', fontsize=12)
plt.ylim(min(total_energy) - 0.002, max(total_energy) + 0.002)
plt.scatter(x=k[2], y=total_energy[2], marker='*', s=150, color='orange', label="Melhor resultado")
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
plt.xticks(k)
plt.legend(fontsize=12, loc='upper right')

# Remover bordas superior e direita
# ax = plt.gca() # Get current axes
# ax.spines['top'].set_visible(False)
# ax.spines['right'].set_visible(False)

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
