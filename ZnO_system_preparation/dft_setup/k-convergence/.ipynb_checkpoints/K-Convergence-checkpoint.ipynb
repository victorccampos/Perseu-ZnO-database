```python
import matplotlib.pyplot as plt
plt.style.use('default')
import numpy as np
```

```python
data = np.loadtxt(fname='dados.dat', delimiter=' ')
```

```python
data
```

```python
k, total_energy, band_gap = data[:,0], data[:,1], data[:,2]
```

```python
# Criar o gráfico

plt.figure(figsize=(8, 5))
plt.scatter(k, total_energy, color='black', s=30, label='Energia total')

# Linha de referência no maior k.
plt.axhline(total_energy[-1], color='gray', linestyle='--', label=f'Referência (k={k[-1]})')

# Rótulos e limites
plt.suptitle('Convergência com número de k-pontos')
plt.xlabel('Número de k-pontos',size=12)
plt.ylabel('Energia total (Ry)', size=12)
plt.ylim(min(total_energy) - 0.002, max(total_energy) + 0.002)
plt.grid(True)
plt.legend()


ax = plt.gca()

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
