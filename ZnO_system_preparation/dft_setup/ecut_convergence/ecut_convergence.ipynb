```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("dados_ecut.dat", sep=' ', skipinitialspace=True, names=['Total Energy', 'Band Gap'])
data
```

```python
data['ecut_values'] = [30, 50, 70, 90, 110, 130, 150]
```

```python
ecutwfc_values = data['ecut_values'].values
energia_total = data['Total Energy'].values
```

```python
# Criar o grfico
plt.figure(figsize=(6, 4))
plt.scatter(ecutwfc_values, energia_total, color='black', s=60, label='Energia total')
plt.axhline(energia_total[-1], color='gray', linestyle='--', label=f'Referncia (ecut={ecutwfc_values[-1]})')

# Remover bordas superior e direita
ax = plt.gca() # Get current axes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Rtulos e limites
plt.title('Convergncia em funo do ecut')
plt.xlabel('ecutwfc', fontsize=12)
plt.xticks(ecutwfc_values)
plt.ylabel('Energia total (Ry)', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=10)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=10)

# Mostrar o grfico
plt.tight_layout()
plt.show()
```
